```python
import pandas as pd

data = {'Name': ['Aryan', 'Dev', 'Yash'],
        'age': [20, 22, 24],
        'City': ['Rajkot', 'Porbandar', 'Junagadh']}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

col_to_del = 'age'

df = df.drop(col_to_del, axis=1)

print("\nDataFrame after deleting the column:")
print(df)
```

```python
import pandas as pd


data = {'Name': ['Aryan', 'Dev', 'Yash'],
        'age': [20, 22, 24],
        'City': ['Rajkot', 'Porbandar', 'Junagadh']}

df = pd.DataFrame(data)

csv_filename = 'output_file.tsv' 
df.to_csv(csv_filename, sep='\t', index=False)

print(f'DataFrame has been written to {csv_filename} with tab separator.')
```
